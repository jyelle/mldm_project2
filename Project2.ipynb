## Read data from csv

```python
import os
import pandas as pd

DATA_DIR = 'data'
CSV_FILE = DATA_DIR + '/KaggleV2-May-2016.csv'

# Read data using read_csv function
def read_csv(data_dir=DATA_DIR, csv_file=CSV_FILE):
    if not os.path.isdir(data_dir):
        os.makedirs(data_dir)
    return pd.read_csv(csv_file)

appt_df = read_csv()
appt_df
```

## Test set creation

- Create a test set of 20k records
- Use stratified sampling on the `No-Show` variable to make sure class proportions are the same for train and test sets
- Save train and test sets as csv files in the `processed_data` directory

```python
PROCESSED_DATA_DIR = 'processed_data'
TRAIN_FILE = PROCESSED_DATA_DIR + '/train_set.csv'
TEST_FILE = PROCESSED_DATA_DIR + '/test_set.csv'

# create test set with stratified sampling here
# refer to the intro notebook posted in class calendar for example
import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=0)
for train_index, test_index in split.split(clean_appt_df, clean_appt_df["No_Show"]):
    train_set = clean_appt_df.loc[train_index]
    test_set = clean_appt_df.loc[test_index]
    
import csv
import os

def save_train_set(processed_data_dir=PROCESSED_DATA_DIR, train_file=TRAIN_FILE):
    if not os.path.isdir(processed_data_dir):
        os.makedirs(processed_data_dir)
    outfile = open(train_file, 'w+')
    outfile.writelines(test_set)
    outfile.close()

def save_test_set(processed_data_dir=PROCESSED_DATA_DIR, test_file=TEST_FILE):
    if not os.path.isdir(processed_data_dir):
        os.makedirs(processed_data_dir)
    outfile = open(test_file, 'w+')
    outfile.writelines(test_set)
    outfile.close()
    
save_train_set()
save_test_set()
```

## Data preprocessing

Create a scikit pipeline for preprocessing the data to prepare it for ML algorithms. At a minimum your pipeline should:

- Standardize numerical variables
- Transform categorical variables into numeric variables

Optionally, you can perform any other transformation you think is suitable. Again, refer to the intro notebook for an example of setting up a preprocessing pipeline.

```python
noshow_data = noshow_data.drop("No_Show", axis=1)
noshow_data_labels = train_set["No_Show"].copy()

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import StandardScaler, LabelBinarizer

cat_attributes = ['Gender','Neighborhood', 'Patient_ID' 'Appointment_ID', 'Scheduled_Day', 'Appointment_Day']
binary_attributes = ['Scholarship', 'Hypertension', 'Diabetes', 'Alcoholism', 'Disability', 'SMS_Received']
num_attributes = [x for x in list(noshow_data) if x not in cat_attributes + binary_attributes]

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        return X[self.attribute_names].values
    
# create pipeline here!
num_pipeline = Pipeline([
    ('selector', DataFrameSelector(num_attributes)),
    #('imputer', Imputer(strategy="median")),
    #('attribs_adder', NormalizedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])

cat_pipeline = Pipeline([
    ('selector', DataFrameSelector(cat_attributes)),
    ('label_binarizer', LabelBinarizer())
])

binary_pipeline = Pipeline([
    ('selector', DataFrameSelector(binary_attributes)),
])

full_pipeline = FeatureUnion(transformer_list=[
    ('num_pipeline', num_pipeline),
    ('cat_pipeline', cat_pipeline),
    ('binary_pipeline', binary_pipeline)
])

# show the first 5 rows of the transformed training set

noshow_data_clean = full_pipeline.fit_transform(noshow_data)
transformed_data = pd.read_csv(noshow_data_clean)
transformed_data.head()
```
